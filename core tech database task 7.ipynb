```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("coretech_clients_cleaned.csv")

df.head()
```

```python
connection = sqlite3.connect("coretech_clients.db")

cursor = connection.cursor()

print("Database Created")
```

```python
create_table = """

CREATE TABLE IF NOT EXISTS clients(

client_id INTEGER,
client_name TEXT,
company_name TEXT,
service TEXT,
project_budget INTEGER,
project_duration_days INTEGER,
client_city TEXT,
lead_source TEXT,
status TEXT,
rating REAL

)

"""

cursor.execute(create_table)

connection.commit()

print("Table Created")
```

```python
df.to_sql(
    "clients",
    connection,
    if_exists="append",
    index=False
)

print("Data Inserted")
```

```python
query = """
SELECT * FROM clients;
"""

result = pd.read_sql(query, connection)

result
```

```python
query = """

SELECT *
FROM clients
WHERE client_city='Lahore';

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT client_name, service, status
FROM clients
WHERE status='Completed';

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT service,
COUNT(*) AS total_clients
FROM clients
GROUP BY service;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT AVG(project_budget)
AS average_budget
FROM clients;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT *
FROM clients
ORDER BY project_budget DESC
LIMIT 1;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT *
FROM clients
ORDER BY project_budget ASC
LIMIT 1;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT client_name, rating
FROM clients
WHERE rating > 4;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT status,
COUNT(*) AS total
FROM clients
GROUP BY status;

"""

pd.read_sql(query, connection)
```

```python
query = """

SELECT client_name,
company_name,
project_budget
FROM clients
ORDER BY project_budget DESC
LIMIT 5;

"""

pd.read_sql(query, connection)
```

```python
connection.close()

print("Database Closed")
```
